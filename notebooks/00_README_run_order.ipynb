# World Cup 2026 forecasting pipeline — START HERE

This is a clean end-to-end notebook chain for the World Cup 2026 forecasting project.
Place all notebooks in one project folder and run them in order.

## Run order

1. `01_build_training_features_from_public_results.ipynb`
   - downloads the public international-results dataset;
   - builds `training_features.csv`;
   - API credits: 0.

2. `02_the_odds_api_historical_odds_fetch.ipynb`
   - uses your The Odds API key;
   - downloads historical odds for World Cup 2022;
   - default budget is about 2,000 credits;
   - creates `historical_odds_aggregated.csv`.

3. `03_join_market_train_and_backtest.ipynb`
   - joins football features with market odds;
   - evaluates market-only, football-only, and football-plus-market baselines;
   - API credits: 0.

4. `04_current_wc2026_odds_ledger.ipynb`
   - builds a daily current-odds ledger for World Cup 2026;
   - API credits: 1–3 per run depending on selected regions.

5. `05_package_project_report_bundle.ipynb`
   - creates one local report bundle with the main outputs.

## Main outputs to inspect

- `data/processed/training_features.csv`
- `data/processed/historical_odds_aggregated.csv`
- `data/processed/current_wc2026_odds_ledger/wc2026_current_odds_ledger.csv`
