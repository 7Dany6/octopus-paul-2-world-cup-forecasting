# 05 — Package project report bundle

Run this after notebooks 01–04.

Output:
- `data/processed/worldcup_chain_full_report_bundle.zip`

This notebook is optional and only packages local outputs for review.


```python
# Cell 1. Package all important outputs.

from pathlib import Path
import zipfile
import pandas as pd
import json

DATA_DIR = Path("data")
PROCESSED_DIR = DATA_DIR / "processed"

zip_path = PROCESSED_DIR / "worldcup_chain_full_report_report_bundle.zip"

include_patterns = [
    "training_features.csv",
    "training_features_with_historical_external.csv",
    "matches_standardized.csv",
    "historical_odds_aggregated.csv",
    "historical_odds/*.csv",
    "historical_odds/*.json",
    "join_train_backtest/*.csv",
    "join_train_backtest/*.json",
    "current_wc2026_odds_ledger/*.csv",
    "current_wc2026_odds_ledger/*.json",
    "01_training_features_report/*.csv",
    "01_training_features_report/*.json",
]

files = []

for pattern in include_patterns:
    files.extend(PROCESSED_DIR.glob(pattern))

files = sorted(set([f for f in files if f.exists() and f.is_file()]))

with zipfile.ZipFile(zip_path, "w", compression=zipfile.ZIP_DEFLATED) as zf:
    for file in files:
        zf.write(file, arcname=str(file.relative_to(PROCESSED_DIR)))

summary = pd.DataFrame([{
    "zip_path": str(zip_path),
    "files_included": len(files),
}])

display(summary)

print("Created:", zip_path.resolve())
print("Report bundle created.")
```
